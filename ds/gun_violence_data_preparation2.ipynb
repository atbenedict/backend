```python
# Generic imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_load = pd.read_csv('gun-violence-data_01-2013_03-2018_cleaned.csv')
```

```python
df_statecode = pd.read_csv('statecode.csv', index_col='state')
df_load['code'] = df_load.apply(lambda row: df_statecode.loc[row.state]['abbreviation'], axis=1)
```

```python
df_load.loc[:,['state','code']]
```

```python
df = df_load[(df_load.state == 'Alaska') |
             (df_load.state == 'Delaware') |
             (df_load.state == 'Louisiana') |
             (df_load.state == 'South Carolina') |
             (df_load.state == 'Illinois') |
             (df_load.state == 'Mississippi') |
             (df_load.state == 'Tennessee') |
             (df_load.state == 'Alabama') |
             (df_load.state == 'Missouri') |
             (df_load.state == 'Maryland')].copy()

df['n_killed_injured'] = df['n_killed'] + df['n_injured']
df = df[df['n_killed_injured'] > 0]
```

```python
df = pd.read_csv('gun-violence-data_01-2013_03-2018_cleaned.csv')
df_statecode = pd.read_csv('statecode.csv', index_col='state')
df['code'] = df.apply(lambda row: df_statecode.loc[row.state]['abbreviation'], axis=1)
df['n_killed_injured'] = df['n_killed'] + df['n_injured']
df_export = df.loc[:,['year', 'state', 'code', 'n_killed_injured']]
```

```python
df_export.to_csv ('gun-violence-data_choropleth.csv', index = False, 
                                   header=True)
```

```python
df_grouped = df_export.groupby('code').sum().reset_index()
```
